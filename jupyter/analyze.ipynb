```python
def 토큰화컬럼을_노멀라이즈(df, 토큰화_컬럼='요구능력'):
    df[토큰화_컬럼] = df[토큰화_컬럼].apply(lambda x: word_tokenize(x) )
    cols = list(df.columns)
    cols.remove(토큰화_컬럼)
    df1 = json_normalize(df.to_dict('records'), 토큰화_컬럼, cols).rename(columns={0:토큰화_컬럼})
    print('\n 토큰화 직후 : ', len(df1))
    return df1
```

```python
def 토큰화컬럼을_정제(df, 토큰화_컬럼):
    df1 = pd.DataFrame({'단어':lh.리스트의_중복제거( list(df[토큰화_컬럼]) )})
    df1.단어 = df1.단어.str.lower()
    df1 = df1.drop_duplicates()
    print('\n 토큰화 단어 필터 후 : ', len(df1))
    
    필요한단어_li = lh.리스트1로부터_리스트2를_제거( list(df1.단어), 특수문자와_선택적품사_리스트() )
    print('\n 선택적 품사 필터 후 : ', len(필요한단어_li))
    return 필요한단어_li
```

```python
# 분석 기초 조작 
df.PFC_가능성 = df.PFC_가능성.apply(lambda x: 'N' if x=='No' else 'Y')

# 토큰화 
토큰화_컬럼='요구능력'
df1 = 토큰화컬럼을_노멀라이즈(df, 토큰화_컬럼)
필요한단어_li = 토큰화컬럼을_정제(df1, 토큰화_컬럼)
기술용어_li = 데이터맵_기술명_가져오기()
df2 = df1[ df1[토큰화_컬럼].isin(기술용어_li) ]
```
